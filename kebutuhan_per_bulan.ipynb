```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sample_data/kebutuhan_per_bulan_2.csv')
```

```python
plt.figure(figsize=(12, 8))  # Mengatur ukuran figure
plt.scatter(df['Propinsi'], df['tahun'], s=50, c='blue', alpha=0.7)

for i in range(len(df)):
    plt.annotate(df['kebutuhan'][i], (df['Propinsi'][i], df['tahun'][i]), ha='center', va='bottom', fontsize=8)

plt.title('Kebutuhan Per Tahun Berdasarkan Provinsi', fontsize=14)
plt.xlabel('Propinsi', fontsize=12)
plt.ylabel('Tahun', fontsize=12)

plt.xticks(rotation=90, fontsize=8)

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
data_filtered = df [df['Propinsi']=='JAWA TIMUR']
data_filtered
```
