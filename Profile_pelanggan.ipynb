```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sample_data/profile_pelanggan.csv')
```

```python
print(df)
```

```python
x = df['Propinsi']
y = df['Data']
plt.figure(figsize=(10, 8))
plt.scatter(x, y)

plt.xlabel('Provinsi')
plt.ylabel('Data')

plt.title('Scatter Plot dari Dataset')

plt.show()
```
