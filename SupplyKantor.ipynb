```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('sample_data/SupplyKantor.csv')
```

```python
print(df)
```

```python
x = df['Kota']
y = df['Jumlah']
plt.figure(figsize=(10, 8))
plt.scatter(x, y)

plt.xlabel('Kota')
plt.ylabel('Jumlah')

plt.title('Jumlah Pesanan dari Setiap Kota')

plt.show()
```

```python
data_filtered = df [df['Kota']=='Bandung']
data_filtered
```

```python
x = data_filtered['Jenis']
y = data_filtered['Jumlah']
plt.figure(figsize=(10, 8))
plt.scatter(x, y)
plt.xlabel('Jenis Barang')
plt.ylabel('Jumlah')

plt.title('Jumlah Pesanan dari Kota Bandung')

plt.show()
```
